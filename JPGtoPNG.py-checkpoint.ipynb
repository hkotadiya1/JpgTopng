# for all png to jpg files

```python
from PIL import Image
import os
```

```python
path = r'C:/Users/JDPC/PIL_Image_Processing'
```

```python
for file in os.listdir(path): 
    if file.endswith(".png"): 
        img = Image.open(file)
        file_name, file_ext = os.path.splitext(file)
        img.save('{}.jpg'.format(file_name))
```
